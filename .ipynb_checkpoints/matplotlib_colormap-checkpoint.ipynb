```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
name = "kekelo"
img = cv2.imread('{}.jpg'.format(name), cv2.IMREAD_UNCHANGED)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
rgb2grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(rgb2grey.shape)
plt.imshow(rgb2grey)
```

```python
plt.imshow(rgb2grey, cmap='plasma')
```
